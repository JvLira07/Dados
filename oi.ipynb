```python
import matplotlib.pyplot as plt
import pandas as pd

# Tente carregar o arquivo com a codificação ISO-8859-1 (ou -15)
df = pd.read_csv('cenario-mortalidade-ceara_2024-09-03_19-25-49.csv', encoding='ISO-8859-1', sep=';')
print(df.columns)

plt.figure(figsize=(12,8))
plt.plot(df['DATA_ÓBITO'],df['ACIDENTES'], label='Acidentes', color='blue')

plt.title('Acidentes por Data de Óbito')
plt.xlabel('Data de Óbito')
plt.ylabel('Número de Acidentes')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
