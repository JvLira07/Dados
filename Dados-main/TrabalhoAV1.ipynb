```python
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv('cenario-mortalidade-ceara_2024-09-03_19-25-49.csv', encoding='ISO-8859-1', sep=';')
print(df.columns)


df_grouped = df.groupby(['ESCOLARIDADE', 'SEXO']).size().unstack()
df_grouped2 = df.groupby(['ACIDENTES', 'SEXO']).size().unstack()
df_grouped3 = df.groupby(['RACA', 'ACIDENTES']).size().unstack()
df_grouped4 = df.groupby(['FAIXA_ETÁRIA', 'ACIDENTES']).size().unstack()


df_grouped.plot(kind='bar', figsize=(12, 8), color=['blue', 'orange','red'], stacked=False)
df_grouped2.plot(kind='bar', figsize=(12, 8), color=['pink', 'blue','red'], stacked=False)
df_grouped3.plot(kind='bar', figsize=(12, 8), color=['pink', 'blue','red'], stacked=False)
df_grouped4.plot(kind='bar', figsize=(12, 8), color=['pink', 'blue','red'], stacked=False)

# Título e rótulos
plt.title('Distribuição por Escolaridade e Sexo')
plt.xlabel('Escolaridade')
plt.ylabel('Contagem')

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()

```
