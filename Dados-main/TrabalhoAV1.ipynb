```python
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv('cenario-mortalidade-ceara_2024-09-03_19-25-49.csv', encoding='ISO-8859-1', sep=';')

df_grouped = df.groupby(['ESCOLARIDADE', 'SEXO']).size().unstack()
df_grouped2 = df.groupby(['ACIDENTES', 'SEXO']).size().unstack()
df_grouped3 = df.groupby(['RACA', 'ACIDENTES']).size().unstack()
df_grouped4 = df.groupby(['FAIXA_ETRIA', 'ACIDENTES']).size().unstack()
df_grouped5 = df.groupby(['NOME_REGIAO', 'ACIDENTES']).size().unstack()

# Grfico 1: Escolaridade e Sexo
ax1 = df_grouped.plot(kind='bar', figsize=(12, 8), color=['blue', 'orange'], stacked=False)
ax1.set_title('Distribuio de Escolaridade por Sexo')
ax1.set_xlabel('Escolaridade')
ax1.set_ylabel('Contagem')
plt.xticks(rotation=45)
plt.tight_layout()

# Grfico 2: Acidentes e Sexo
ax2 = df_grouped2.plot(kind='bar', figsize=(12, 8), color=['pink', 'blue'], stacked=False)
ax2.set_title('Distribuio de Acidentes por Sexo')
ax2.set_xlabel('Nmero de Acidentes')
ax2.set_ylabel('Contagem')
plt.xticks(rotation=45)
plt.tight_layout()

# Grfico 3: Raa e Acidentes
ax3 = df_grouped3.plot(kind='bar', figsize=(12, 8), color=['purple', 'green'], stacked=False)
ax3.set_title('Distribuio de Raa por Nmero de Acidentes')
ax3.set_xlabel('Raa')
ax3.set_ylabel('Contagem')
plt.xticks(rotation=45)
plt.tight_layout()

# Grfico 4: Faixa Etria e Acidentes
ax4 = df_grouped4.plot(kind='bar', figsize=(12, 8), color=['red', 'yellow'], stacked=False)
ax4.set_title('Distribuio de Faixa Etria por Nmero de Acidentes')
ax4.set_xlabel('Faixa Etria')
ax4.set_ylabel('Contagem')
plt.xticks(rotation=45)
plt.tight_layout()

# Grfico 5: Regio e Acidentes
ax5 = df_grouped5.plot(kind='bar', figsize=(12, 8), color=['brown', 'cyan'], stacked=False)
ax5.set_title('Distribuio de Acidentes por Regio')
ax5.set_xlabel('Regio')
ax5.set_ylabel('Contagem')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
